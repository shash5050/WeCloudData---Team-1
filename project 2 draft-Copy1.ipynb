```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
df['StockCode'].nunique()
```

```python
df=df.loc[df['Quantity']>0]
```

```python
pd.DataFrame(df.groupby('StockCode')['Description'].count())
```

```python
df.StockCode.count()
```

```python
df_pivot_table=pd.pivot_table(df,index='StockCode',columns='InvoiceDate',values='Quantity',aggfunc='sum')
```

```python
df_pivot_table.head()
```

```python
df_pivot_table=df_pivot_table.applymap(lambda x: 1 if x>0 else 0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
df_cosin_similarity=cosine_similarity (df_pivot_table)
```

```python
# consine_similarity package will pairwise each sample if there is only one parameter inside
df_cosin_similarity=pd.DataFrame(df_cosin_similarity)
```

```python
df_cosin_similarity.head()
```

```python
df_cosin_similarity=df_cosin_similarity.set_index(df_pivot_table.index)
df_cosin_similarity
```

```python
# top 10 similary items for StockCode=16219
df_cosin_similarity.loc[16219,:].sort_values(ascending=False).head(11)
```

```python
top_10_items=list(df_cosin_similarity.loc[16219,:].sort_values(ascending=False).head(11).index)
```

```python
top_10_items.pop(0)
```

```python
top_10_items
```

```python
df.loc[df['StockCode'].isin(top_10_items),['StockCode','Description']]
```

```python
#there are so many relicas for items and description as well, for better visulization just drop duplicates
df.loc[df['StockCode'].isin(top_10_items),['StockCode','Description']].drop_duplicates()
```

Conclusion: For Customer who viewed a good with StockCode 16219, we'd like to recommend these 10 items to him/her.
